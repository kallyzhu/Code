```python
import time 
import datetime
import numpy as np
import pandas as pd
```

```python
#apply to simple sample
df = pd.DataFrame({'A': [-0.1,0.3,np.nan,-0.4, 0.8,-0.5],'B': [-0.2,0.5,0.3,-0.5,np.nan,0.1],'C':['2019/03/03','2019/06/04','2019/09/03','2019/12/05','2020/03/04','2020/06/06']},index=[0, 1, 2, 3,4,5])

#set index
df['C']=df['C'].apply(lambda x: pd.Timestamp(x))
df.set_index(['C'],inplace=True)
df
```

```python
#rolling by day
df.rolling('730d', min_periods=1).sum()
```

```python
#rolling by window size
df.rolling(4, min_periods=1).sum()
```

```python
#with self-designed function 
#import data and transfer date into timestamp
df=pd.read_csv('test_data.csv')
df.head()
```

```python
df['DataDate']=df['DataDate'].apply(lambda x: pd.Timestamp(x))
df=df[['DataDate', 'SALE', 'SALEq']]
df.isna().sum()
```

```python
#apply function:
#Here I want to sum the most recent 4 records and if there is missing values, rooling back 2 years at most,
#i.e. don't return values 2 years ago.
#############################################################
def sum_four_quuater(x):
    df=pd.DataFrame(x)
    df=df.dropna()
    return df.tail(4).sum()
df.set_index('DataDate').rolling('730d',min_periods=1).apply(sum_four_quuater)
```
